```python
# categorical and numeric variables after x and y seperation
categorical_variables= [features for features in x.columns if x[features].dtype=='object']
numerical_variables= [features for features in x.columns if x[features].dtype in(['int64','float'])]
numerical_variables
categorical_variables
```

```python
# create column transformer with 3 types of transformers 
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer


# numeric column transformer 
numeric_transformer=StandardScaler() 
# using one hot encoder to convert categorical faetures into numerical features  
one_hot=OneHotEncoder()

preprocessor = ColumnTransformer(

    [
        ("OneHotEncoder",one_hot,categorical_variables),
        ('StandardScaler',numeric_transformer,numerical_variables)

    ]
)
```

```python
# fit _transform the x - independent variables 
# fitting the model
x=preprocessor.fit_transform(x)
```
